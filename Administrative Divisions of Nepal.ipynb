```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/Nepal'
headers = {
    "User-Agent": "Bot/1.0 (https://example.com) Python requests"
}
page = requests.get(url, headers=headers)
```

```python
soup=BeautifulSoup(page.text,'html.parser')
# print(soup)
```

```python
table=soup.find_all('table')[1]
# print(table)
```

```python
headings=table.find_all('th')[0:8]
print(headings)
```

```python
titles=[title.text for title in headings]
print(titles)
```

```python
titles=[title.text.strip() for title in headings]
print(titles)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=titles)
df
```

```python
data=table.find_all('tr')[1:8]
# data
```

```python
for info in data:
    row=info.find_all('td')[:8]
    ind_row=[d.text.strip()  for d in row]
    length=len(df)
    df.loc[length]=ind_row



last_row=[i.text.strip() for i in table.find_all('tr')[8]]
last_row[1:17:2]
length=len(df)
df.loc[length]=last_row[1:17:2]

    
```
